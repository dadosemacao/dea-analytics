# Teorema do Limite Central e sua aplicação

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parâmetros da população exponencial
media_populacao = 10  # Média da população
tamanho_amostra = 100  # Tamanho da amostra
numero_amostras = 1000  # Número de amostras a serem retiradas
```

```python
# Gerando amostras e calculando médias
medias_amostrais = []
for _ in range(numero_amostras):
    amostra = np.random.exponential(scale=media_populacao, size=tamanho_amostra)
    media_amostral = np.mean(amostra)
    medias_amostrais.append(media_amostral)
```

```python
# Plotando histograma das médias amostrais
plt.hist(medias_amostrais, bins=30, density=True, alpha=0.6, color='g')

# Plotando densidade de probabilidade normal teórica
sigma = media_populacao / np.sqrt(tamanho_amostra)
x = np.linspace(min(medias_amostrais), max(medias_amostrais), 100)
pdf_normal = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(- (x - media_populacao)**2 / (2 * sigma**2))
plt.plot(x, pdf_normal, color='r', linewidth=2)

plt.title('Histograma das Médias Amostrais vs. Distribuição Normal Teórica')
plt.xlabel('Médias Amostrais')
plt.ylabel('Densidade de Probabilidade')
plt.legend(['Distribuição Normal Teórica', 'Médias Amostrais'])
plt.show()
```

**Exemplo:**
vou criar um exemplo simples em Python que demonstra como o Teorema do Limite Central pode ser utilizado para monitorar e controlar a qualidade em um processo de fabricação. Vamos supor que estamos monitorando o peso de produtos em uma linha de produção e queremos garantir que a média dos pesos esteja dentro das especificações.

```python
# Parâmetros do processo de fabricação
mu = 100  # Peso médio alvo
sigma = 2  # Desvio padrão do processo

# Tamanho da amostra
n = 30  # Número de produtos em cada amostra

# Número de amostras
num_samples = 1000
```

```python
# Definindo limites de controle (e.g., ±3 desvios padrão da média)
limite_inferior = mu - 3 * sigma
limite_superior = mu + 3 * sigma

print(f"limite_inferior: {limite_inferior}")
print(f"limite_superior: {limite_superior}")
```

```python
# Simulando pesos de produtos em uma linha de produção
pesos = np.random.normal(mu, sigma, size=(num_samples, n))

# Calculando as médias de cada amostra
media_amostral = np.mean(pesos, axis=1)
```

```python
# Verificando se as médias estão dentro dos limites de controle
for i, media in enumerate(media_amostral):
    if media < limite_inferior or media > limite_superior:
        print(f"Amostra {i+1}: Média = {media:.2f} fora dos limites de controle.")
```

```python
# Calculando a porcentagem de amostras dentro dos limites de controle
dentro_limites = np.mean((media_amostral >= limite_inferior) & (media_amostral <= limite_superior)) * 100
print(f"\nPorcentagem de amostras dentro dos limites de controle: {dentro_limites:.2f}%")
```
